```python
import pandas as pd
import os
```

```python
parent_directory = os.path.dirname('./')
```

```python
filepath = os.path.join(parent_directory, 'source-data', 'Kevin', "Kevin's Publication TCRs Updated.xlsx")
```

```python
df_tcr = pd.read_excel(filepath, sheet_name='TCRs from Tetramer Signals')
df_tcr.head()
```

```python
df_antigen = pd.read_excel(filepath, sheet_name='Antigen_sequences', skiprows=[0])
df_antigen.head()
```

```python
# apply inner join the TCR table and antigen table on the antigen name. ignored cross reactivity
df_merge = pd.merge(df_tcr, df_antigen, how='inner', left_on='specificity.rename', right_on='Antigen')
df_merge.head()
```

```python
df_result = df_merge.drop(['Cell_ID', 'TCRa_1st_cdr_nt', 'TCRb_1st_cdr_nt', 'TCR', 'tcr.frequency', 'specificity.rename', 'gene.final', 'Unnamed: 0', 'Antigen', 'Gene', 'Group'], axis=1)
df_result.columns
```

```python
df_result.rename({'TCRa_1st_cdr_aa': 'CDR3.alpha.aa', 'TCRa_1st_v': 'V.alpha', 'TCRa_1st_j': 'J.alpha', 'TCRb_1st_cdr_aa': 'CDR3.beta.aa',
       'TCRb_1st_v': 'V.beta', 'TCRb_1st_j': 'J.beta', 'AA sequence of antigen': 'Epitope.peptide', 'MHC': 'Epitope.peptide'}, axis=1, inplace=True)
```

```python
df_result.head()
```

```python
# remove TCR if the CDR sequence contains any uncertain amino acid
import re
pattern = re.compile(r"[^ARNDCEQGHILKMFPSTWYV]")
illegals = []
for index, row in df_result.iterrows():
    if pattern.search(row['CDR3.alpha.aa']) or pattern.search(row['CDR3.beta.aa']):
        illegals.append(index)
df_result.drop(illegals, axis=0, inplace=True)
```

```python
len(df_result['Epitope.peptide'].unique())
```

```python
df_result['HLA'] = df_result['HLA'].apply(lambda x: f'HLA-{x[0]}*{x[1:3]}:{x[3:]}')
```

```python
df_result['V.alpha'] = df_result['V.alpha'].apply(lambda x: x.split('*')[0])
df_result['J.alpha'] = df_result['J.alpha'].apply(lambda x: x.split('*')[0])
df_result['V.beta'] = df_result['V.beta'].apply(lambda x: x.split('*')[0])
df_result['J.beta'] = df_result['J.beta'].apply(lambda x: x.split('*')[0])
```
